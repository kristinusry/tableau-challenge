```python
# Dependencies
import pandas as pd
```

```python
# Jan
Citibike_Jan_df = "JC-201901-citibike-tripdata.csv"
Citibike_Jan_df = pd.read_csv(Citibike_Jan_df)
# Feb
Citibike_Feb_df = "JC-201902-citibike-tripdata.csv"
Citibike_Feb_df = pd.read_csv(Citibike_Feb_df)
#March
Citibike_Mar_df = "JC-201903-citibike-tripdata.csv"
Citibike_Mar_df = pd.read_csv(Citibike_Mar_df)
#April
Citibike_Apr_df = "JC-201904-citibike-tripdata.csv"
Citibike_Apr_df = pd.read_csv(Citibike_Apr_df)
#May
Citibike_May_df = "JC-201905-citibike-tripdata.csv"
Citibike_May_df = pd.read_csv(Citibike_May_df)
#June
Citibike_Jun_df = "JC-201906-citibike-tripdata.csv"
Citibike_Jun_df = pd.read_csv(Citibike_Jun_df)
#July
Citibike_Jul_df = "JC-201907-citibike-tripdata.csv"
Citibike_Jul_df = pd.read_csv(Citibike_Jul_df)
#August
Citibike_Aug_df = "JC-201908-citibike-tripdata.csv"
Citibike_Aug_df = pd.read_csv(Citibike_Aug_df)
#Sept
Citibike_Sep_df = "JC-201909-citibike-tripdata.csv"
Citibike_Sep_df = pd.read_csv(Citibike_Sep_df)
#Oct
Citibike_Oct_df = "JC-201910-citibike-tripdata.csv"
Citibike_Oct_df = pd.read_csv(Citibike_Oct_df)
#Nov
Citibike_Nov_df = "JC-201911-citibike-tripdata.csv"
Citibike_Nov_df = pd.read_csv(Citibike_Nov_df)
#Dec
Citibike_Dec_df = "JC-201912-citibike-tripdata.csv"
Citibike_Dec_df = pd.read_csv(Citibike_Dec_df)
```

```python
Citibike_2019_df = pd.concat([Citibike_Jan_df, 
                              Citibike_Feb_df, 
                              Citibike_Mar_df, 
                              Citibike_Apr_df, 
                              Citibike_May_df, 
                              Citibike_Jun_df, 
                              Citibike_Jul_df, 
                              Citibike_Aug_df, 
                              Citibike_Sep_df, 
                              Citibike_Oct_df, 
                              Citibike_Nov_df, 
                              Citibike_Dec_df], axis=0)
Citibike_2019_df
```

```python
Citibike_2019_df['gender'] = Citibike_2019_df['gender'].map({1: 'Male', 2: 'Female',0: 'Unknown'})
Citibike_2019_df
```

```python
Citibike_2019_df.to_csv('Citibike_2019_data.csv', index=False)
```
